```python
# Dynamic Data Storage in Python Using Lists and Dictionaries

student = []

field = list(map(str, input("Enter the Field names separated by space").split()))

num_student = int(input("Enter number of student you want :"))

for i in range(num_student):
    print(f"Enter the detail of {i+1} student")
    dic = {}

    for j in field:
        value = str(input(f"Enter the {j}: "))
        dic[j] = value

    student.append(dic)

print(student)        
```
