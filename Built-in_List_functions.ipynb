**Bulit-in Functions of List**

```python
list1 = [1,2,3,4,5,6]

# For length
print(len(list1)) 

# For sum of all the elements
print(sum(list1)) 

# Max and min of list
print(max(list1)) 
print(min(list1)) 

# Sorting the list
list2 = [5,2,4,7,9,3]
print(sorted(list2))

print(sorted(list2, reverse = True)) # Reverse Sorting

# Reverse the list
list3 = [5,6,7,8,9,10]
list3.reverse() # Permanent changing list to reverse order
print(list3)

new_list = list(reversed(list3))
print(new_list)    # reversed copy

# Best way for printing reverse list
print(list1[::-1])

# Count function
list4 = [2,4,2,3,2,2,2]
print(list4.count(2))

# Copy list
list5 = list4.copy()
print(list5)

# Clear list
list5.clear()
print(list5)

# Extend list
list6 = [10,20]
list4.extend(list6)
print(list4)        # Know the difference between Append and extend
```

**List comprehension**

```python
number = [1,2,3,4,5]

sqaure = [n*n for n in number]

print(number)
print(sqaure)
```

```python
number = [1,2,3,4,5,6]

evens = [n for n in number if n % 2 ==0]

print(number)
print(evens)
```

```python
numbers = [1,2,3,4,5,6]

result = [n if n%2 == 0 else "Odd" for n in numbers]

print(numbers)
print(result)
```

```python
sqaure = [x*x for x in range(5)]

print(sqaure)
```

```python
numbers = [1,2,2,3,3,4,4]

unique = {n for n in numbers}
print(unique)
```
