```python
# Dynamic List + Dictionary

students = []

fields = input("Enter the field name separated by space: ").split()
student_count = int(input("Enter the number of students you want: "))

for i in range(student_count):
    student = {}
    print(f"\nEnter values for student {i + 1}:")
    
    for field in fields:
        student[field] = input(f"Enter {field}: ")
        
    students.append(student)


# Print Full list 
def print_list(students, fields):
    for stu in students:

        for field in fields:
            print(f"{field}: {stu[field]}")

        print("------------")
    

# Print all the data of particular field
def field_list(students, fields):
    field_name = input("Enter the field name: ")

    if field_name not in fields:
        print("Invalid Field Name")
        return

    for stu in students:
        print(stu[field_name])

        
# Find student by value of any field
def find_stu(students, fields):
    value = input("Enter the value to search: ")
    found = False

    for stu in students:
        if value in stu.values():
            found = True
            for field in fields:
                print(f"{field}: {stu[field]}")
            print("-------")

    if not found:
        print(f"{value} not found")

        
# Count no of students of perticular field
def count_stu(students):
    value = input("Enter the value to count: ")
    count = sum(1 for stu in students if value in stu.values())

    if count == 0:
        print(f"{value} not found")
    else:
        print(f"There are {count} students with value '{value}'.")

        
# Add new Dictionary
def add_dic(students, fields):
    student = {}
    for field in fields:
        student[field] = input(f"Enter {field}: ")
    students.append(student)


# Delete whole dictionary by value of field name 
def delete_dic(students):
    value = input("Enter value to delete student: ")
    original_len = len(students)

    students[:] = [stu for stu in students if value not in stu.values()]

    if len(students) == original_len:
        print(f"{value} not found")


# Update the information by student name
def update_dic(students, fields):
    if "id" not in fields:
        print("No 'id' field exists.")
        return

    value = input("Enter the id to update: ")

    for stu in students:
        if stu["id"] == value:
            for field in fields:
                if field != "id":
                    stu[field] = input(f"Enter new {field}: ")
            return

    print("Student not found")


while True:
    choice = int(input(
        "\033[1m Enter the operation you want to perform: \033[0m\n"
        "1. Print Full list\n"
        "2. Print all the data of particular field\n"
        "3. Find student by value of any field\n"
        "4. Count no of students of particular field\n"
        "5. Add new Dictionary\n"
        "6. Delete whole dictionary by value of field name\n"
        "7. Update the information by student name\n"
        "8. Exit\n"
    ))
    
    match choice:
        case 1:
            print_list(students, fields)
        case 2:
            field_list(students, fields)
        case 3:
            find_stu(students, fields)
        case 4:
            count_stu(students)
        case 5:
            add_dic(students, fields)
        case 6:
            delete_dic(students)
        case 7:
            update_dic(students, fields)
        case 8: 
            print("Exiting program...")
            break
        case _:
            print("Invalid choice")
```
