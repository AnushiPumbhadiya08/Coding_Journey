**4. Loops**

```python
# For Loops
print("1. For loop")     
for i in range(5):
    print(i)
    
print("\n")

print("2.For loop for list")
fruits = ["Apple 12" , "Banana 01" , "Cherry"]

for j in fruits:
    print(j)

print("\n")

print("3.Loop through a string")
for k in "Python":
    print(k)

print("\n")

print("4. Using range(start, stop, step)")
for l in range(1, 10, 2):
    print(l)
```

```python
#Function and While loop
def print_number(n):
    m = 1
    while n >= m:
        print(m)
        m += 1

no = int(input("Enter the number: "))
print_number(no)
        
```

```python
# Some Functions of list

number = [0,1,2,3]

number[0] = 100   # Changing value
print("Printing after changing value =",number)

number.append(4)  # Adding number at last
print("Printing after adding number at last =",number)

number.insert(2,200) # Interting value at given index
print("Printing after inserting number at given index =",number)

number.remove(200) # Removing given value
print("Printing after removing given number =",number)

number.pop(3) # Removing by index
print("printing after removing number from given index =",number)

print(len(number)) # For printing length of list
```

```python
# Mini Exercise
# Make function for sum of number in a list

def sum_list(number):
    total = 0
    for i in number:
        total += i
    return total

number = [1,2,3,4,5]
result = sum_list(number)
print(result)
```

```python
# Make dynamic list 

number = []

n = int(input("How many total number do you need ?"))

for i in range(n):
    value = int(input(f"Enter the {i+1} Element: "))
    number.append(value)   # Try other functions.

print(number)
```

```python
# Another way to make list

numbers = list(map(int, input("Enter numbers separated by space: ").split()))
print(numbers)
```
